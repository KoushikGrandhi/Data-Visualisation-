```
import pandas as pd
import plotly.express as px
```

```
df= pd.read_csv('data.csv')
```

```
df.info()
```

```
df = df.dropna().reindex()
```

```
df.head()
```

```
import plotly.graph_objects as go
```

```
df= df.sort_values('Year',axis=0,ascending=True)
df[df.columns[-1]] = df[df.columns[-1]] *100
```

```
fig = px.choropleth(df,locations="Country Name",
    locationmode='country names',color='Proportion of seats held by women in national parliaments (%)',projection='equirectangular',
        hover_name="Country Name",
        hover_data=['Year','Proportion of seats held by women in national parliaments (%)'],
      color_continuous_scale=px.colors.sequential.Plasma,
       title = '% Women at influential positions',
      labels={"Proportion of seats held by women in national parliaments (%)": "% of women in parliaments"},
     animation_frame='Year'
      )
fig.update_layout(
       
        showlegend = True,
        geo = dict(
            landcolor = 'rgb(200, 200, 200)',
            showcountries=True
        ),
         margin=dict(l=00, r=00, t=15
                     , b=0)
    )
fig.update_layout(
  title_font_size=22,
    title={
        'y':0.99,
        'x':0.35,
        'xanchor': 'center',
        'yanchor': 'top'}
)
```

```
df_new =  df.groupby(['Year'])['Proportion of seats held by women in national parliaments (%)'].sum()
```

```
import plotly.io as pio

aggs = ["avg","stddev","min","max","first","last"]

agg = []
agg_func = []
for i in range(0, len(aggs)):
    agg = dict(
        args=['transforms[0].aggregations[0].func', aggs[i]],
        label=aggs[i],
        method='restyle'
    )
    agg_func.append(agg)
```

```
data = [dict(
  type = 'choropleth',
  locationmode = 'country names',
  locations = df['Country Name'],
  z = df['Proportion of seats held by women in national parliaments (%)'],
  autocolorscale = False,
  colorscale = 'Mint',
  reversescale = True,
  transforms = [dict(
    type = 'aggregate',
    groups = df['Country Name'],
    aggregations = [dict(
        target = 'z', func = 'sum', enabled = True)
    ]
  )]
)]
```

```
layout = dict(
  title = '<b>Women at Influential positions</b><br>Use dropdown to change aggregation (All data  in %)',
  legend='%',
  showlegend=False,
  xaxis = dict(title = 'Subject'),
  yaxis = dict(title = 'Score', range = [0,22]),
  # height = 600,
  # width = 900,
   margin=dict(l=00, r=00, t=15
                     , b=0),
  updatemenus = [dict(
        x = 0.75,
        y = 1.10,
        xref = 'paper',
        yref = 'paper',
        yanchor = 'top',
        active = 1,
        showactive = False,
        buttons = agg_func
  )]
)
```

```
fig_dict = dict(data=data, layout=layout)

pio.show(fig_dict, validate=False)
```
